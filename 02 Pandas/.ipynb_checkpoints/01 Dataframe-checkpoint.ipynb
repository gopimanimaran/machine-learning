```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
# select subset of colums to apply; apply the "apply"  function and then pass the columns, mention axis to make sure it applies correct on columns
df['Other'] = df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

```python
df
```

```python
df['Other'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df
```

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''
  
# code snippet whose execution time is to be measured 
stmt_one = ''' 
df['Other'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Other'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
  
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_one, 
                    number = 1000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_two, 
                    number = 1000) 
```
