```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/data_banknote_authentication.csv')
```

```python
df.isnull().sum()
```

```python
X=df.drop('Class',axis=1)
```

```python
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
help(RandomForestClassifier)
```

```python
n_estimators = [64,100,128,200]
max_features=[2,3,4]
bootstrap=[True,False]
oob_score=[True,False]
param_grid = {'n_estimators':n_estimators,'max_features':max_features,'bootstrap':bootstrap,'oob_score':oob_score}
```

```python
model = RandomForestClassifier()
```

```python
grid = GridSearchCV(model,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
model_actual = RandomForestClassifier(max_features= 2, n_estimators= 100, oob_score= True)
```

```python
model_actual.fit(X_train,y_train)
```

```python
y_pred = model_actual.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
