```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/rock_density_xray.csv')
```

```python
df
```

```python
df.columns = ['Signal','Density']
```

```python
df
```

```python
sns.scatterplot(x='Signal',y='Density',data=df)
```

```python
X = df['Signal'].values.reshape(-1,1) #reshape is needed
y = df['Density']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

```python
y_pred
```

```python
signal_range = np.arange(1,100)
signal_range
```

```python
signal_pred = model.predict(signal_range.reshape(-1,1))
signal_pred
```

```python
sns.scatterplot(x='Signal',y='Density',data=df)
plt.plot(signal_range,signal_pred)
#Linear regression is not able to fit into curvy
```

```python
#Lets create a function to perform Fit model,metrics and plot graph
def run_model(model,X_train, X_test, y_train, y_test):
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    mae = mean_absolute_error(y_test,preds)
    print(f'MAE and rmse are : {mae} and {rmse}')
    signal_range = np.arange(1,100)
    signal_pred = model.predict(signal_range.reshape(-1,1))
    plt.figure(figsize=(12,8),dpi=200)
    sns.scatterplot(x='Signal',y='Density',data=df,color='black')
    plt.plot(signal_range,signal_pred)
```

```python
model = LinearRegression()
run_model(model,X_train, X_test, y_train, y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
pipe = make_pipeline(PolynomialFeatures(degree=6),LinearRegression())
```

```python
run_model(pipe,X_train, X_test, y_train, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
k_values = [1,2,5,10,20]

for n in k_values:
    model=KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train, X_test, y_train, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()

run_model(model,X_train, X_test, y_train, y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr = SVR()
param_grid = {'C':[0.01,0.1,1,5,10,100,1000],'gamma':['scale','auto']}

grid = GridSearchCV(svr,param_grid)
run_model(grid,X_train, X_test, y_train, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=10)
run_model(model,X_train, X_test, y_train, y_test)
```
