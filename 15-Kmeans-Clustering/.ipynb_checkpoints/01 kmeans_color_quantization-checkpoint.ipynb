```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```python
#Read image as an array
image_as_arr = mpimg.imread('../DATA/palm_trees.jpg')
```

```python
image_as_arr
```

```python
#H,W and Color channels RGB
image_as_arr.shape
```

```python
plt.imshow(image_as_arr)
```

```python
image_as_arr.shape
```

```python
(h,w,c) = image_as_arr.shape
```

```python
h
```

```python
#KMeans need 2D array, lets unpack 3D into 2D
image_as_array2d = image_as_arr.reshape(h*w,c)
```

```python
image_as_array2d
```

```python
image_as_array2d.shape
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
#lets make 2D to 3D
quantized_pic = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_pic
```

```python
plt.imshow(quantized_pic)
```
