```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head(5)
```

```python
X=df.drop('sales',axis=1)
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polinomial_convertor = PolynomialFeatures(degree=2,include_bias=False)
```

```python
polinomial_convertor.fit(X)
```

```python
poly_features = polinomial_convertor.transform(X)
```

```python
poly_features.shape
```

```python
X.iloc[0]
```

```python
poly_features[0]
#a,b,c (3)
#a,b,c,ab,bc,ca,a*a,b*b,c*c (9)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predictions = model.predict(X_test)
```

```python
model.coef_
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
MAE
```

```python
MSE = mean_squared_error(y_test,test_predictions)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
#How to decide which degree we should we o with onpolynomial
#See error vs degree diagram
```

```python
train_rmse_errors=[]
test_rmse_errors=[]
for d in range(1,10):
    poly_conv = PolynomialFeatures(degree=d,include_bias=False)
    poly_feat = poly_conv.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(poly_feat, y, test_size=0.3, random_state=101)
    
    model = LinearRegression()
    model.fit(X_train,y_train)
    
    train_pred=model.predict(X_train)
    test_pred=model.predict(X_test)
    
    train_rmse = np.sqrt(mean_squared_error(y_train,train_pred)) 
    test_rmse = np.sqrt(mean_squared_error(y_test,test_pred)) 
    
    train_rmse_errors.append(train_rmse)
    test_rmse_errors.append(test_rmse)
```

```python
train_rmse_errors
```

```python
test_rmse_errors
```

```python
plt.plot(range(1,6),train_rmse_errors[:5],label='TRAIN')
plt.plot(range(1,6),test_rmse_errors[:5],label='TEST')
plt.legend();
```

```python
final_poly_conv = PolynomialFeatures(degree=3,include_bias=False)
```

```python
final_poly_feat = final_poly_conv.fit_transform(X)
```

```python
final_model=LinearRegression()
final_model.fit(final_poly_feat,y)
```

```python
from joblib import load,dump
```

```python
dump(final_model,'poly_final.joblib')
```

```python
dump(final_poly_conv,'final_poly_conv.joblib')
```
