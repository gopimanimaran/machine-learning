```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head(5)
```

```python
sns.pairplot(data=df)
```

```python
X=df.drop('sales',axis=1)
```

```python
X
```

```python
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(X_test)
```

```python
len(X_train)
```

```python
len(df)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predictions = model.predict(X_test)
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
mean_absolute_error(y_test,test_predictions)
```

```python
mean_squared_error(y_test,test_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
test_residuals = y_test - test_predictions
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='red')
```

```python
final_model=LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
final_model.coef_
#This gives the values for how each X feature is going to affect for each unit (TC, radio and newspaper)
```

```python
y_hat = final_model.predict(X)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].plot(df['TV'],y_hat,'o',color='red')
axes[0].set_ylabel("TV Sales")
axes[0].set_title("TV spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].plot(df['radio'],y_hat,'o',color='red')
axes[1].set_ylabel("Sales")
axes[1].set_title("TV spend")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].plot(df['newspaper'],y_hat,'o',color='red')
axes[2].set_ylabel("newspaper Sales")
axes[2].set_title("newspaper spend")
plt.tight_layout();
```

```python
from joblib import dump,load
dump(final_model,'final_sale_model.joblib')
```
