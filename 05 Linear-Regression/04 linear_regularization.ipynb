```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head(5)
```

```python
X=df.drop('sales',axis=1)
y=df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polinomial_convertor = PolynomialFeatures(degree=3,include_bias=False)
```

```python
polinomial_features = polinomial_convertor.fit_transform(X)
```

```python
polinomial_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(polinomial_features, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#Fit only on Traning data set
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
polinomial_features[0]
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train,y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
```

```python
MAE
```

```python
RMSE=np.sqrt(mean_squared_error(y_test,test_predictions))
RMSE
```

```python
ridge_cv_model.fit(X_train,y_train) #only used X_train and out of train data sets some are used for testing k times
```

```python
ridge_cv_model.alpha_
```

```python
from sklearn.metrics import SCORERS
SCORERS.keys()
```
